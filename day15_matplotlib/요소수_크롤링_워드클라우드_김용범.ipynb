```python
import requests # 서버를 위한 request
from bs4 import BeautifulSoup # 크롤링을 위한 bs4
import pandas as pd
```

```python
source = requests.get('https://www.yna.co.kr/view/AKR20211111016800003?input=1195m')
soup = BeautifulSoup(source.content, "html.parser")
```

```python
table = soup.find('article',{'class':'story-news article'})
```

```python
article = table.find_all('p')[2:-5]
```

```python
text = ''
for i in article:
    text += (i.text)
```

```python
text
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wc = WordCloud(background_color='white', random_state=0,font_path='AppleGothic')
plt.figure(figsize=(20,15))
plt.imshow(wc.generate(text))
plt.axis('off')
plt.show()
```
