### 결측값 제거

```python
# 행의 결측값 제거 : 변수.dropna(axis=0)
# 열의 결측값 제거 : 변수.dropna(axis=1)
df.info()
```

```python
df.dropna() # 변수 df의 데이터 프레임 중 결측값(NaN) 제거
```

```python
df.dropna(axis=0) # 행 제거
```

```python
df.dropna(axis=1) # 열 제거
```

```python
df.dropna(axis=0,how='any') # 한개라도 있는 경우
```

```python
df.dropna(axis=0,how='all') # 모두 NaN인 경우 
```

```python
df.dropna(axis=1,how='all')
```

```python
# 데이터 프레임 합치기 : merge
# merge 명령은 두 데이터프레임이 공통으로 가지고 있는 열 또는 인덱스를 기준으로 두 테이블을 합칠 때 사용한다.
# append는 행 기준
cust = pd.DataFrame({
    '고객번호' : [1001,1002,1003,1004,1005,1006,1007],
    '이름' : ['가','나','다','라','마','바','사']
}, columns=['고객번호','이름'])
cust
```

```python
order = pd.DataFrame({
    '고객번호' : [1001,1001,1005,1006,1008,1001],
    '금액' : [10000,20000,15000,5000,100000,30000]
}, columns=['고객번호','금액'])
order
```

```python
pd.merge(cust,order) # 두 테이블 합치기 서로 비교해서 없으면 나오지 않음
```

```python
pd.merge(cust,order,how='outer')
```

```python
pd.merge(cust,order,how='left') #왼쪽에 해당하는 cust가 모두 나옴
```

```python
order
```

```python
pd.merge(cust,order,how='right')
```
