```python
def getDate(data):
    data = str(data)
    year = data[0:4]
    month = data[4:6]
    day = data[6:8]
    return year + '-' + month + '-' + day
```

```python
d = 20210201
getDate(d)
```

```python
df['사용일자1'] = df['사용일자'].apply(getDate)
df.head()
```
