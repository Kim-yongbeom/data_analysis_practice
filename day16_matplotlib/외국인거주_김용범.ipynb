```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font',family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
sns.set(font="AppleGothic", 
        rc={"axes.unicode_minus":False})
```

```python
df = pd.read_csv('5개시도_인구데이터(외국인).csv')
df
```

```python
sns.barplot(x=df['sido'],y=df['foreigner_all(2018)'])
```

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(x=df['foreigner_all(2017)'],y=df['sgg'])
```

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(x=df['foreigner_all(2018)'],y=df['sgg'])
```

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(x=df['foreigner_all(2019)'],y=df['sgg'])
```

```python
sns.boxplot(x=df['sido'],y=df['foreigner_all(2018)'])
```

```python
sns.histplot(x=df['sido'],y=df['foreigner_all(2019)'])
```

```python
sns.set_style('darkgrid')
sns.set(font="AppleGothic", 
        rc={"axes.unicode_minus":False})
fig = plt.figure(figsize=(17, 5))   
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)
sns.barplot(x=df['sido'],y=df['foreigner_all(2017)'],ax=ax1)
sns.barplot(x=df['sido'],y=df['foreigner_all(2018)'],ax=ax2)
sns.barplot(x=df['sido'],y=df['foreigner_all(2019)'],ax=ax3)
```

- 분석
- 1) 조사한 5개의 도시에서의 외국인은 점차 늘어나고 있는 경향을 보인다.
- 2) 서울과 경기도의 평균값이 거의 비슷하지만 서울 특정 지역에서 매우 높은 이상치를 보여주고 있어 전체적인 외국인의 수는 높다 (구로구, 영등포구)
