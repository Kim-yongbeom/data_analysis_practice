```python
import json
import folium
from folium import plugins
import pandas as pd
```

```python
subway = pd.read_csv('./지하철_위경도_처리.csv',encoding='cp949')
subway
```

```python
line1 = subway[subway.선명=='1호선']
line1.reset_index(drop=True, inplace=True)
line1
```

```python
line2 = subway[subway.선명=='2호선']
line2.reset_index(drop=True, inplace=True)
line2
```

```python
map = folium.Map(location = [37.52389,126.96586], tiles='OpenStreetMap',zoom_start=11)

# 각 지하철역을 마커를 사용하여 지도에 표시하기
for name, lat, long in zip(line2.역이름, line2.위도, line2.경도):
    folium.Marker([lat,long], tooltip=name).add_to(map)

# 지도 전체화면
plugins.Fullscreen(position='topright', title='Click to Expand', title_cancel='Click to Exit', force_separate_button=True).add_to(map)

# 지도 스크롤
plugins.MousePosition().add_to(map)

# 지도 띄우기
map
```
