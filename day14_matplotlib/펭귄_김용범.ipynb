```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('penguins')
df
```

```python
df.describe()
```

```python
x = df['bill_length_mm']
y = df['bill_depth_mm']
sns.scatterplot(x=x,y=y,hue=df['species'])
```

```python
x = df['body_mass_g']
y = df['bill_depth_mm']
sns.scatterplot(x=x,y=y,hue=df['sex'])
```

```python
pivot_df = df.pivot_table('body_mass_g','bill_length_mm','bill_depth_mm')
sns.heatmap(pivot_df,cmap='Blues')
```

```python
sns.lineplot(x=df['bill_depth_mm'],y=df['body_mass_g'],ci=100)
```

```python
sns.lineplot(x=df['bill_depth_mm'],y=df['body_mass_g'],hue=df['island'],ci=None)
```
