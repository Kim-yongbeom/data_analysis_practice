```python
# 기본 모듈
from selenium import webdriver
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
```

```python
# 맬론 동적으로 들어가기
driver = webdriver.Chrome("chromedriver")
sleep(0.5)
driver.get('https://www.melon.com/chart/index.htm')
soup = BeautifulSoup(driver.page_source)
soup
```

```python
melon_rank = soup.find_all('span',{'class':'rank'})[1:]
melon_rank
```

```python
for i in melon_rank:
    melon_song_rank = i.string
    print(melon_song_rank)
```

```python
# 제목 태그 전체 검색
melon_title = soup.select('div.rank01 > span > a')
melon_title
```

```python
# 제목 한줄씩 출력
for i in melon_title:
    malon_song_title = i.text
    print(malon_song_title)
```

```python
# 가수 태그 전체 출력
malon_singer = soup.select("div.rank02")
malon_singer
```

```python
# 가수 전체 출력
for i in malon_singer:
    malon_song_singer = i.span.text
    print(malon_song_singer)
```

```python
# 딕셔너리 안에 반복문해줘서 한줄씩 출력
melon_Top100 = []
for i in range(100):
    melon_song_rank = melon_rank[i].text
    malon_song_title = melon_title[i].text
    malon_song_singer = malon_singer[i].span.text
    melon_Top100.append(melon_song_rank)
    melon_Top100.append(malon_song_title)
    melon_Top100.append(malon_song_singer)
```

```python
melon_Top100
```

```python
driver = webdriver.Chrome("chromedriver")
sleep(0.5)
driver.get('https://www.genie.co.kr/chart/top200')
soup = BeautifulSoup(driver.page_source)
soup
```

```python
for i in range(1,51):
    genie_song_rank = i
    print(genie_song_rank)
```

```python
genie_title = soup.find_all('a',{'class':'title ellipsis'}) # select, findAll 번갈아가며 써보는중
genie_title
```

```python
for i in genie_title:
    genie_song_title = i.text.strip() # 줄간격을 지니에서 이상하게 해놔서 strip으로 정렬해줌
    print(genie_song_title)
```

```python
genie_singer = soup.find_all('a',{'class':'artist ellipsis'})
genie_singer
```

```python
for i in genie_singer:
    genie_song_singer = i.text
    print(genie_song_singer)
```

```python
# 한 페이지에 50개밖에 없으므로 top 50
genie_Top50 = []
for i in range(50):
    genie_song_rank = i+1
    genie_song_title = genie_title[i].text.strip()
    genie_song_singer = genie_singer[i].text
    genie_Top50.append(genie_song_rank)
    genie_Top50.append(genie_song_title)
    genie_Top50.append(genie_song_singer)
```

```python
genie_Top50
```

```python
driver = webdriver.Chrome("chromedriver")
sleep(0.5)
driver.get('https://music.bugs.co.kr/chart')
soup = BeautifulSoup(driver.page_source)
soup
```

```python
for i in range(1,101):
    bugs_song_rank = i
    print(bugs_song_rank)
```

```python
bugs_title = soup.find_all('p',{'class':'title'})
bugs_title
```

```python
for i in bugs_title:
    bugs_song_title = i.a.string
    print(bugs_song_title)
```

```python
bugs_singer = soup.find_all('p',{'class':"artist"})
bugs_singer
```

```python
for i in bugs_singer:
    bugs_song_singer = i.a.text
    print(bugs_song_singer)
```

```python
bugs_Top100 = []
for i in range(100):
    bugs_song_rank = i
    bugs_song_title = bugs_title[i].a.string
    bugs_song_singer = bugs_singer[i].a.text
    bugs_Top100.append(bugs_song_rank)
    bugs_Top100.append(bugs_song_title)
    bugs_Top100.append(bugs_song_singer)
```

```python
bugs_Top100
```

```python
melon_result = []
for i in range(100):    
    melon_result.append([melon_rank[i].text]+[melon_title[i].text]+[malon_singer[i].span.text])

melon_tbl = pd.DataFrame(melon_result, columns=("순위", "곡 명", "가수"))
print(melon_tbl)
```

```python
genie_result = []
for i in range(50):    
    genie_result.append([i+1]+[genie_title[i].text.strip()]+[genie_singer[i].text])

genie_tbl = pd.DataFrame(genie_result, columns=("순위", "곡 명", "가수"))
print(genie_tbl)
```

```python
bugs_result = []
for i in range(100):    
    bugs_result.append([i+1]+[bugs_title[i].a.string]+[bugs_singer[i].a.text])

bugs_tbl = pd.DataFrame(bugs_result, columns=("순위", "곡 명", "가수"))
print(bugs_tbl)
```

```python
with pd.ExcelWriter('MUSIC_CHART.xlsx') as writer:
    melon_tbl.to_excel(writer, encoding='utf-8', sheet_name='멜론차트')
    genie_tbl.to_excel(writer, encoding='utf-8', sheet_name='지니차트')
    bugs_tbl.to_excel(writer, encoding='utf-8', sheet_name='벅스차트')
```
