# 전체 사이트를 합산하여 가장 인기가 있는 가수 10팀을 구하시오. 

```python
# 일단 찾을 부분의 regex문을 정의해준다 
# 괄호를 사용하여 괄호 안의 값을 찾는다 (str) 
# re모듈을 사용할 때 패턴을 만들어서 사용할 수 있다. 
# 패턴을 미리 컴파일하면, 여러 번 재사용할 때 편하기 때문이다. 
# 패턴을 명시할 때, r 문자를 사용하는 것을 볼 수있다. 

# \(, \) < 특수기호 앞에는 백슬래쉬를 넣어서 지정해준다 
# . < 아무 문자/기호/스페이스 
# + < 1개 이상 
pd_genie.가수.str.replace(r'\(.+\)','')

# 리턴 값을 확인해보니 가수이름에 공백이 포함된것들이 있다.
# 공백제거를 한 값들을 리턴해본다 
a = pd_genie.가수.str.replace(r'\(.+\)', '')
a = a.str.strip()
a
# 확인 완료 -> 값을 assign해준다 
```

```python
# 확인 완료 -> 값을 assign해준다 
# (표기) 제거 
pd_genie['가수'] = pd_genie.가수.str.replace(r'\(.+\)', '', regex=True)
pd_genie['가수']  = pd_genie['가수'].str.strip()

pd_bugs['가수'] = pd_bugs.가수.str.replace(r'\(.+\)', '', regex=True)
pd_bugs['가수'] = pd_bugs.가수.str.strip()

pd_melon['가수'] = pd_melon.가수.str.replace(r'\(.+\)', '', regex=True)
pd_melon['가수'] = pd_melon.가수.str.strip()

display(pd_genie)
display(pd_bugs)
display(pd_melon)
```

```python
# 가수값만 새로운 df에 저장 
# df 생성 
team = pd.DataFrame() 

team=team.append(pd_bugs[['가수']])
team=team.append(pd_melon[['가수']])
team=team.append(pd_genie[['가수']])

pd.DataFrame(team.가수.value_counts().head(15))
```
