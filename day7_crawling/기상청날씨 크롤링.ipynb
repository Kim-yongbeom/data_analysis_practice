```python
import requests # 서버를 위한 request
from bs4 import BeautifulSoup # 크롤링을 위한 bs4
import pandas as pd
```

```python
source = requests.get('https://www.weather.go.kr/weather/observation/currentweather.jsp')
soup = BeautifulSoup(source.content, "html.parser")
```

```python
table = soup.find('table',{'class':'table_develop3'})
data = []
```

```python
print('='*30)
print('\n       오늘의 기상 예보입니다.\n')
print('='*30)
# [code 0]
for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))
    for td in tds:
        if td.find('a'):
            location = td.find('a').text # 지역
            temp = tds[5].text # 온도
            humidity = tds[9].text # 습도
#             print(location, temp, humidity)
            print("{0:<7} {1:<7} {2:<7}".format(location, temp, humidity)) # 위에 print와 같다
            data.append([location, temp, humidity])
print('='*30)
print('\n       좋은 하루 보내세요!\n')
print('='*30)
```

```python
with open('weather.csv','w') as f:
    f.write('지역, 온도, 습도\n')
    for i in data:
        f.write('{0},{1},{2}\n'.format(i[0],i[1],i[2]))
df = pd.read_csv('weather.csv', index_col='지역', encoding='utf-8')
city_df = df.loc[['서울','인천','대전','대구','광주','부산','울산']]
print(city_df)
```
