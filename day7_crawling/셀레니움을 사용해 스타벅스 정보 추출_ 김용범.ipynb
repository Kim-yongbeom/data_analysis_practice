```python
# 데이터 구조
import pandas as pd
import numpy as np

# 웹 크롤링을 위하여
from selenium import webdriver
from bs4 import BeautifulSoup

# 크롤링 통신을 위하여
import requests
import json
from pandas.io.json import json_normalize

# 웹드라이버 의 Select 와 key 를 위하여
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys

# 기타
import time
```

```python
# 웹 드라이버 실행
driver=webdriver.Chrome("chromedriver")
```

```python
def click(area):
    driver.get("https://www.starbucks.co.kr/store/store_map.do?disp=locale")
    time.sleep(5)
    driver.implicitly_wait(30) # 다음 트리거 까지 최대 30초 기다리기 그 전에 완료되면 상관없음
    
    ss = driver.find_element_by_class_name("sido_arae_box")
    s1 = ss.find_elements_by_tag_name('li')
    time.sleep(3)
    driver.implicitly_wait(10)
    
    ##############################################
    
    s1[area].click()
    time.sleep(3)
    driver.implicitly_wait(30)
    
    ##############################################
    
    gg = driver.find_element_by_class_name('gugun_arae_box')
    g1 = gg.find_elements_by_tag_name('li')
    time.sleep(2)
    driver.implicitly_wait(10)
    g1[0].click()
    
    ##############################################
    
    # 로딩 대기
    time.sleep(5)
    driver.implicitly_wait(30)
    
    # 소스코드
    source = driver.page_source
    bs = BeautifulSoup(source,'lxml')
    return bs
```

```python
def getInfo(bs):
    ul = bs.find('ul', class_='quickSearchResultBoxSidoGugun')
    sb_list = ul.find_all('li')
    time.sleep(3)
    driver.implicitly_wait(30)
    
    # 매장의 정보를 담을 리스트 생성
    names = []
    lat = []
    long = []
    add = []
    # 각 매장의 정보를 받아온다
    for sb in sb_list:
        names.append(sb['data-name']) # sb 안에 있는 data-name을 가져오는것 (코드 안에 있음)
        lat.append(sb['data-lat'])
        long.append(sb['data-long'])
        add.append(sb.find('p').text)
        
    # 데이터 프레임 생성
    area = pd.DataFrame([names, lat, long, add], index=['매장이름','위도','경도','주소']).T
    return area
```

```python
# 실행!!
# 서울 = 0, 경기 = 1
bs = click(0)    # li 0번째는 서울
ss_sb = getInfo(bs)
bs = click(1)    # li 1번째는 경기
gg_sb = getInfo(bs)
```
