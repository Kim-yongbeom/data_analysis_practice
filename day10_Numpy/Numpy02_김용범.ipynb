```python
import numpy as np
points = np.array([[1,1],[1,2,],[1,3],[2,1],[2,2],[2,3],[3,1],[3,2],[3,3]])
p = np.array([2.5,2])
import matplotlib.pyplot as plt
plt.plot(points[:,0],points[:,1],"ro")
plt.plot(p[0],p[1],'bo')
```

```python
x = np.linspace(0,5,10)
y = x**2
plt.plot(x,y);
```

```python
x = np.linspace(0,10,20)
y = x**2.0
plt.plot(x,y,'bo-',linewidth=3,markersize=5) # bo- : 점에 선을 연결 , bo-- : 점선
```

```python
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x,y1,"bo-", linewidth=2, markersize=5,label='Frist')
plt.plot(x,y2,"gs-", linewidth=2, markersize=5,label='Second')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([-0.5,10.5,-5,105]) # 왼쪽부터 2개는 x축 , 나머지 2개는 y축
plt.legend(loc='upper left')
plt.savefig('mplot.pdf')
plt.savefig('mplot.jpg')
plt.savefig('mplot.png')
```

```python
x = np.random.standard_normal(size=1000)
plt.hist(x) # 표준정규분포
```

```python
plt.hist(x, density=True) # 빈도수가 아닌 확률로 나타내고 싶으면 hist안에 density=True 입력
```
